```python
import numpy as np
import os
from PIL import Image, ImageDraw, ImageFont
import matplotlib.pyplot as plt
from google import genai
from pydantic import BaseModel
```

```python
gemini_api_key = os.environ.get('GEMINI_API_KEY')
client = genai.Client(api_key=gemini_api_key)
```

```python
class ObjectBB(BaseModel):
    bounding_boxes: list[float]
    object_name: str
```

```python
img = Image.open("desk.jpg")
# img = img.rotate(-90)
width, height = img.size

plt.imshow(img)
plt.axis("off")
plt.show()
```

```python
PROMPT = ("Provide the bounding boxes for any and all objects in this image that an object detector may be interested in using the [ymin, xmin, ymax, xmax] format.")
```

```python
# Ask the model to generate the bounding boxes
response = client.models.generate_content(
model="gemini-1.5-flash",
contents=[img, PROMPT],
config={
        'response_mime_type': 'application/json',
        'response_schema': list[ObjectBB],
        }
)
```

```python
bboxes_and_names: list[ObjectBB] = response.parsed
```

```python
img = Image.open("desk.jpg")
# img = img.rotate(-90)

# Define a list of colors to cycle through
colors = ["red", "blue", "green", "yellow", "purple", "orange", "violet"]


draw = ImageDraw.Draw(img)
for entry in bboxes_and_names:
    box = entry.bounding_boxes
    name = entry.object_name

    y1, x1, y2, x2 = box
    y1 = y1/1000 * height
    x1 = x1/1000 * width
    y2 = y2/1000 * height
    x2 = x2/1000 * width

    # Get the color for the current bounding box
    color = colors[bboxes_and_names.index(entry) % len(colors)]

    # Draw the bounding box on the image
    draw.rectangle([(x1, y1), (x2, y2)], outline=color, width=20)

    # Define the font and size
    font = ImageFont.truetype("arial.ttf", size=150)

    # Write the object name above the bounding box with the specified font
    draw.text((x1, y1-150), name, fill=color, font=font)

plt.imshow(img)
plt.axis("off")
plt.show()
```
